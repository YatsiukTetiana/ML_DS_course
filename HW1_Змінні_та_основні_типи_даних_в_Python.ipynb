**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4
```

```python
print("Addition:", var1 + var2)
print("Subtraction:", var1 - var2)
print("Multiplication:", var1 * var2)
print("Division:", var1 / var2)
print("Integer Division:", var1 // var2)
print("Modulus:", var1 % var2)
print("Exponentiation:", var1 ** var2)
```
