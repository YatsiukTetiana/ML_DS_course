1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x_vals = np.linspace(0,10,100)
```

```python
# Symbolic variable
x = sp.symbols('x')

# Linear function
f_lin = x*10 - 5
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції:", derivative_f_lin)
```

```python
df_vals = np.full_like(x_vals, derivative_f_lin)
f_lin = x_vals*10 - 5
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_lin, 'o', label="f(x)=10x - 5")
plt.plot(x_vals, df_vals, label="f'(x)=10")
plt.title("Графік функції")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

plt.show()
```

Висновок:
Похідна функції є сталим значенням (10), отже, з визначення похідної, "швидкість" зміни оцінки з кожною годиною навчання становить 10 балів. Кожна додаткова година навчання додає 10 балів

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Symbolic variable
x = sp.symbols('x')

# sin(x) function
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

# ln(x) function
f_ln = sp.log(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна функції sin(x):", derivative_f_sin)
print("Похідна функції ln(x):", derivative_f_ln)
```

```python
x_lin = np.linspace(0,10,100)

f_sinx = np.sin(x_lin)
der_f_sinx = np.cos(x_lin)

f_lnx = np.log(x_lin)
der_f_lnx = 1/x_lin
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x_lin, f_sinx, label="f(x)=sin(x)")
plt.plot(x_lin, der_f_sinx, label = "f'(x)=cos(x)")

plt.title("Графік функції")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x_lin, f_lnx, label="f(x)=ln(x)")
plt.plot(x_lin, der_f_lnx, label="f'(x)=1/x")
plt.title("Графік функції")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Symbolic variable
x = sp.symbols('x')

# Function
f_spendings = x**2 - 5*x + 1
derivative_f_spendings = sp.diff(f_spendings, x)

print("Похідна функції sin(x):", derivative_f_spendings)
```

```python
plt.figure(figsize=(10,5))

plt.plot(x_range, f_sp, label = 'f(x) = x^2 - 5x + 1')
plt.plot(x_range, der_f_sp, label = 'f`(x) = 2x - 5')

plt.title("Графік функції витрат")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

plt.show()
```
