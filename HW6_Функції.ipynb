**1. Функція з параметром**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
def show_menu(dishes):
    print("=== Меню ===")
    for i, dish in enumerate(dishes):
        print(f"{i+1}. {dish}")
    print("=============")

show_menu(["Суп грибний", "Котлета по-київськи", "Лимонад"])
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
    return f"Квадрат числа {n} = {n ** 2}"

print(print_square(5))
print(print_square(10))
print(print_square(13))
````

````python
result1 = (lambda x: f"Квадрат числа {x} = {x ** 2}")(5)
result2 = (lambda x: f"Квадрат числа {x} = {x ** 2}")(10)
result3 = (lambda x: f"Квадрат числа {x} = {x ** 2}")(13)
print(result1)
print(result2)
print(result3)
````

**3. Функція з поверненням значення**

Створіть функцію `multiply(a, b)`, яка повертає добуток двох чисел.

Викличте функцію на власноруч заданих значеннях та виведіть результат.


````python
def multiply(a, b):
    return a * b

print(multiply(4, 5))
print(multiply(7, 3))
print(multiply(10, 2))
````

````python
multiply1 = (lambda a, b: a * b)(4, 5)
multiply2 = (lambda a, b: a * b)(7, 3)
multiply3 = (lambda a, b: a * b)(10, 2) 
print(multiply1)
print(multiply2)
print(multiply3)
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
def find_max(numbers):
    return max(numbers)

numbers = [15, 42, 8, 23, 56, 19]
max_number = find_max(numbers)
print(f"Максимальне число в списку: {max_number}")
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
def check_grade(grade):
    if grade >= 60:
        return "Здав"
    else:
        return "Не здав"

print(check_grade(75))
print(check_grade(45))
print(check_grade(60))
````

````python
check_grade1 = (lambda grade: "Здав" if grade >= 60 else "Не здав")(75)
check_grade2 = (lambda grade: "Здав" if grade >= 60 else "Не здав")(45)
check_grade3 = (lambda grade: "Здав" if grade >= 60 else "Не здав")(60)
print(check_grade1)
print(check_grade2)
print(check_grade3)
````
