1. Задані дві матриці `A` і `B`розмірностями 3x2 та 2х3. Виконайте їхнє множення, щоб отримати матрицю 2x2. Виведіть результуючу матрицю `result_matrix` на екран.

3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

1. Для того, щоб отримати розмірність 2х2, потрібно перемножити B*A

```python
result_matrix = np.dot(B, A)

print("Оригінальна матриця A", A, sep='\n')
print("\nОригінальна матриця B", B, sep='\n')
print("\nРезультат множення матриць", result_matrix, sep='\n')
```

2. Множення матриці на вектор  

Якщо result_matrix має розмірність 2х2, а вектор С має 2х1, то результатом множення буде вектор розмірністю 2х1 (відповідно до правила розмірностей при мнрженні матриць та векторів:  
Amхn Bnхp = Cmхp

```python
result_matrix_C = np.dot(result_matrix, C)
print("Результат множення на вектор С", result_matrix_C, sep='\n')
```

3. Множення матриць 2х2 одна на одну  
При такому множенні результатом буде матриця такої ж розмірності

```python
result_matrix_D = np.dot(result_matrix, D)
print("Результат множення на D", result_matrix_D, sep='\n')
```
